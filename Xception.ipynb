```python
# ============================================================
# ğŸ§  Xception Classification Brain Tumor Dataset (Extraction + Fine-Tuning)
# ============================================================

import os
import zipfile
import pickle
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report

# ============================================================
# 1ï¸âƒ£ DÃ©tection GPU / CPU
# ============================================================
if tf.config.list_physical_devices('GPU'):
    print("âœ… GPU dÃ©tectÃ© â€” utilisation de l'accÃ©lÃ©ration matÃ©rielle")
else:
    print("âš ï¸ Aucun GPU dÃ©tectÃ© â€” exÃ©cution sur CPU")

# ============================================================
# 2ï¸âƒ£ Extraction automatique du ZIP
# ============================================================
zip_path = "/content/archive (2).zip"
extract_path = "/content/brain_dataset"
os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
print(f"âœ… ZIP extrait dans : {extract_path}")

# ============================================================
# 3ï¸âƒ£ DÃ©tection automatique du sous-dossier si nÃ©cessaire
# ============================================================
subfolders = [f for f in os.listdir(extract_path) if os.path.isdir(os.path.join(extract_path, f))]
if len(subfolders) == 1:
    extract_path = os.path.join(extract_path, subfolders[0])
    print(f"ğŸ“‚ Chemin corrigÃ© automatiquement vers : {extract_path}")

# ============================================================
# 4ï¸âƒ£ Chemins vers Training / Testing
# ============================================================
train_dir = os.path.join(extract_path, "Training")
test_dir = os.path.join(extract_path, "Testing")

if not os.path.exists(train_dir) or not os.path.exists(test_dir):
    raise FileNotFoundError(f"âŒ Impossible de trouver Training ou Testing dans {extract_path}")

# ============================================================
# 5ï¸âƒ£ PrÃ©paration des donnÃ©es
# ============================================================
IMG_SIZE = 299  # Taille recommandÃ©e pour Xception
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# ============================================================
# 6ï¸âƒ£ DÃ©finition du modÃ¨le Xception
# ============================================================
base_model = tf.keras.applications.Xception(
    weights='imagenet',
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False
)
base_model.trainable = False  # GÃ¨le les couches initiales

def get_pretrained(num_classes):
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = tf.keras.applications.xception.preprocess_input(inputs)
    x = base_model(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(512, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    return keras.Model(inputs=inputs, outputs=outputs)

model = get_pretrained(train_generator.num_classes)
model.summary()

# ============================================================
# 7ï¸âƒ£ Compilation du modÃ¨le initial
# ============================================================
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ============================================================
# 8ï¸âƒ£ Callbacks
# ============================================================
early_stop = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
checkpoint = ModelCheckpoint('xception_brain_tumor.h5', monitor='val_loss', save_best_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)

# ============================================================
# 9ï¸âƒ£ EntraÃ®nement initial
# ============================================================
epochs_initial = 20

history = model.fit(
    train_generator,
    validation_data=test_generator,
    epochs=epochs_initial,
    callbacks=[early_stop, checkpoint, reduce_lr]
)

# ============================================================
# ğŸ”Ÿ Fine-Tuning
# ============================================================
base_model.trainable = True
for layer in base_model.layers[:-20]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

epochs_finetune = 30

history_finetune = model.fit(
    train_generator,
    validation_data=test_generator,
    epochs=epochs_finetune,
    callbacks=[early_stop, checkpoint, reduce_lr]
)

# ============================================================
# 1ï¸âƒ£1ï¸âƒ£ Ã‰valuation sur test set
# ============================================================
loss, acc = model.evaluate(test_generator)
print(f"\nâœ… Test Accuracy: {acc*100:.2f}%")
print(f"âœ… Test Loss: {loss:.4f}")

# ============================================================
# 1ï¸âƒ£2ï¸âƒ£ MÃ©triques avancÃ©es
# ============================================================
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes

precision = precision_score(y_true, y_pred_classes, average='weighted')
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')
accuracy = accuracy_score(y_true, y_pred_classes)

print("\nğŸ“Š MÃ©triques globales sur le jeu de test :")
print(f"   â€¢ PrÃ©cision (Precision) : {precision*100:.2f}%")
print(f"   â€¢ Rappel (Recall)       : {recall*100:.2f}%")
print(f"   â€¢ F1-Score              : {f1*100:.2f}%")
print(f"   â€¢ Exactitude (Accuracy) : {accuracy*100:.2f}%")

# Rapport par classe
labels = list(test_generator.class_indices.keys())
report = classification_report(y_true, y_pred_classes, target_names=labels)
print("\nğŸ“„ Rapport dÃ©taillÃ© par classe :\n")
print(report)

# ============================================================
# 1ï¸âƒ£3ï¸âƒ£ Sauvegarde de lâ€™historique
# ============================================================
with open("xception_history.pkl", "wb") as f:
    pickle.dump({"initial": history.history, "finetune": history_finetune.history}, f)
print("ğŸ’¾ Historique sauvegardÃ© avec succÃ¨s.")

# ============================================================
# 1ï¸âƒ£4ï¸âƒ£ Courbes dâ€™apprentissage
# ============================================================
def plot_history(histories, metric, title):
    plt.figure(figsize=(8,6))
    for label, history in histories.items():
        plt.plot(history[metric], label=f'{label} {metric}')
        plt.plot(history[f'val_{metric}'], label=f'{label} val_{metric}')
    plt.title(title)
    plt.xlabel('Ã‰poques')
    plt.ylabel(metric)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_history({"initial": history.history, "finetune": history_finetune.history}, 'accuracy', "ğŸ“ˆ Courbe de PrÃ©cision (Accuracy)")
plot_history({"initial": history.history, "finetune": history_finetune.history}, 'loss', "ğŸ“‰ Courbe de Perte (Loss)")

# ============================================================
# 1ï¸âƒ£5ï¸âƒ£ Matrice de confusion
# ============================================================
cm = confusion_matrix(y_true, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
plt.figure(figsize=(8,6))
disp.plot(cmap='Blues', xticks_rotation=45)
plt.title("ğŸ§© Matrice de Confusion â€” Xception")
plt.show()
```
