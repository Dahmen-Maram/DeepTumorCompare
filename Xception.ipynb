```python
# ============================================================
# 🧠 Xception Classification Brain Tumor Dataset (Extraction + Fine-Tuning)
# ============================================================

import os
import zipfile
import pickle
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint, ReduceLROnPlateau
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report

# ============================================================
# 1️⃣ Détection GPU / CPU
# ============================================================
if tf.config.list_physical_devices('GPU'):
    print("✅ GPU détecté — utilisation de l'accélération matérielle")
else:
    print("⚠️ Aucun GPU détecté — exécution sur CPU")

# ============================================================
# 2️⃣ Extraction automatique du ZIP
# ============================================================
zip_path = "/content/archive (2).zip"
extract_path = "/content/brain_dataset"
os.makedirs(extract_path, exist_ok=True)

with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)
print(f"✅ ZIP extrait dans : {extract_path}")

# ============================================================
# 3️⃣ Détection automatique du sous-dossier si nécessaire
# ============================================================
subfolders = [f for f in os.listdir(extract_path) if os.path.isdir(os.path.join(extract_path, f))]
if len(subfolders) == 1:
    extract_path = os.path.join(extract_path, subfolders[0])
    print(f"📂 Chemin corrigé automatiquement vers : {extract_path}")

# ============================================================
# 4️⃣ Chemins vers Training / Testing
# ============================================================
train_dir = os.path.join(extract_path, "Training")
test_dir = os.path.join(extract_path, "Testing")

if not os.path.exists(train_dir) or not os.path.exists(test_dir):
    raise FileNotFoundError(f"❌ Impossible de trouver Training ou Testing dans {extract_path}")

# ============================================================
# 5️⃣ Préparation des données
# ============================================================
IMG_SIZE = 299  # Taille recommandée pour Xception
batch_size = 32

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=True,
    fill_mode='nearest'
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# ============================================================
# 6️⃣ Définition du modèle Xception
# ============================================================
base_model = tf.keras.applications.Xception(
    weights='imagenet',
    input_shape=(IMG_SIZE, IMG_SIZE, 3),
    include_top=False
)
base_model.trainable = False  # Gèle les couches initiales

def get_pretrained(num_classes):
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = tf.keras.applications.xception.preprocess_input(inputs)
    x = base_model(x)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(512, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    return keras.Model(inputs=inputs, outputs=outputs)

model = get_pretrained(train_generator.num_classes)
model.summary()

# ============================================================
# 7️⃣ Compilation du modèle initial
# ============================================================
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# ============================================================
# 8️⃣ Callbacks
# ============================================================
early_stop = EarlyStopping(monitor='val_loss', patience=8, restore_best_weights=True)
checkpoint = ModelCheckpoint('xception_brain_tumor.h5', monitor='val_loss', save_best_only=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=3, min_lr=1e-6)

# ============================================================
# 9️⃣ Entraînement initial
# ============================================================
epochs_initial = 20

history = model.fit(
    train_generator,
    validation_data=test_generator,
    epochs=epochs_initial,
    callbacks=[early_stop, checkpoint, reduce_lr]
)

# ============================================================
# 🔟 Fine-Tuning
# ============================================================
base_model.trainable = True
for layer in base_model.layers[:-20]:
    layer.trainable = False

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

epochs_finetune = 30

history_finetune = model.fit(
    train_generator,
    validation_data=test_generator,
    epochs=epochs_finetune,
    callbacks=[early_stop, checkpoint, reduce_lr]
)

# ============================================================
# 1️⃣1️⃣ Évaluation sur test set
# ============================================================
loss, acc = model.evaluate(test_generator)
print(f"\n✅ Test Accuracy: {acc*100:.2f}%")
print(f"✅ Test Loss: {loss:.4f}")

# ============================================================
# 1️⃣2️⃣ Métriques avancées
# ============================================================
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes

precision = precision_score(y_true, y_pred_classes, average='weighted')
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')
accuracy = accuracy_score(y_true, y_pred_classes)

print("\n📊 Métriques globales sur le jeu de test :")
print(f"   • Précision (Precision) : {precision*100:.2f}%")
print(f"   • Rappel (Recall)       : {recall*100:.2f}%")
print(f"   • F1-Score              : {f1*100:.2f}%")
print(f"   • Exactitude (Accuracy) : {accuracy*100:.2f}%")

# Rapport par classe
labels = list(test_generator.class_indices.keys())
report = classification_report(y_true, y_pred_classes, target_names=labels)
print("\n📄 Rapport détaillé par classe :\n")
print(report)

# ============================================================
# 1️⃣3️⃣ Sauvegarde de l’historique
# ============================================================
with open("xception_history.pkl", "wb") as f:
    pickle.dump({"initial": history.history, "finetune": history_finetune.history}, f)
print("💾 Historique sauvegardé avec succès.")

# ============================================================
# 1️⃣4️⃣ Courbes d’apprentissage
# ============================================================
def plot_history(histories, metric, title):
    plt.figure(figsize=(8,6))
    for label, history in histories.items():
        plt.plot(history[metric], label=f'{label} {metric}')
        plt.plot(history[f'val_{metric}'], label=f'{label} val_{metric}')
    plt.title(title)
    plt.xlabel('Époques')
    plt.ylabel(metric)
    plt.legend()
    plt.grid(True)
    plt.show()

plot_history({"initial": history.history, "finetune": history_finetune.history}, 'accuracy', "📈 Courbe de Précision (Accuracy)")
plot_history({"initial": history.history, "finetune": history_finetune.history}, 'loss', "📉 Courbe de Perte (Loss)")

# ============================================================
# 1️⃣5️⃣ Matrice de confusion
# ============================================================
cm = confusion_matrix(y_true, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=labels)
plt.figure(figsize=(8,6))
disp.plot(cmap='Blues', xticks_rotation=45)
plt.title("🧩 Matrice de Confusion — Xception")
plt.show()
```
