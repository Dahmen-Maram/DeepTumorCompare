```python
# ============================================================
# üß† ResNet152V2 Classification Brain Tumor Dataset (GPU auto)
# ============================================================

import os
import zipfile
import pickle
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import precision_score, recall_score, f1_score, accuracy_score, classification_report

# ============================================================
# 1Ô∏è‚É£ Configuration GPU
# ============================================================
print("üîç P√©riph√©riques disponibles :")
print(tf.config.list_physical_devices())

# ============================================================
# 2Ô∏è‚É£ Extraction du ZIP
# ============================================================
zip_path = "/content/archive (2).zip"        # üîÑ Remplace par ton chemin exact
extract_path = "/content/brain_dataset"

os.makedirs(extract_path, exist_ok=True)

if zipfile.is_zipfile(zip_path):
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(extract_path)
    print(f"‚úÖ Extraction termin√©e : {extract_path}")
else:
    print("‚ùå Fichier ZIP invalide.")
    exit()

# Corrige le chemin si un dossier interm√©diaire existe
subfolders = os.listdir(extract_path)
if len(subfolders) == 1 and os.path.isdir(os.path.join(extract_path, subfolders[0])):
    extract_path = os.path.join(extract_path, subfolders[0])
    print(f"üìÇ Chemin corrig√© vers : {extract_path}")

# ============================================================
# 3Ô∏è‚É£ Pr√©paration des donn√©es
# ============================================================
train_dir = os.path.join(extract_path, "Training")
test_dir  = os.path.join(extract_path, "Testing")

IMG_SIZE = 224
batch_size = 16

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(
    train_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=batch_size,
    class_mode='categorical'
)

test_generator = test_datagen.flow_from_directory(
    test_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=batch_size,
    class_mode='categorical',
    shuffle=False
)

# ============================================================
# 4Ô∏è‚É£ Mod√®le ResNet152V2
# ============================================================
base_model = tf.keras.applications.ResNet152V2(
    weights='imagenet',
    include_top=False,
    input_shape=(IMG_SIZE, IMG_SIZE, 3)
)
base_model.trainable = False

def get_pretrained(num_classes):
    inputs = layers.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
    x = base_model(inputs)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.Dropout(0.1)(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    return keras.Model(inputs, outputs)

model = get_pretrained(train_generator.num_classes)

# ============================================================
# 5Ô∏è‚É£ Compilation et callbacks
# ============================================================
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
checkpoint = ModelCheckpoint('resnet152v2_brain_tumor.h5', monitor='val_loss', save_best_only=True)

# ============================================================
# 6Ô∏è‚É£ Entra√Ænement
# ============================================================
epochs = 50
history = model.fit(
    train_generator,
    validation_data=test_generator,
    epochs=epochs,
    callbacks=[early_stop, checkpoint]
)

# Sauvegarde de l'historique
with open("history.pkl", "wb") as f:
    pickle.dump(history.history, f)

print("‚úÖ Mod√®le et historique sauvegard√©s.")

# ============================================================
# 7Ô∏è‚É£ √âvaluation finale + m√©triques avanc√©es
# ============================================================
print("\nüîé √âvaluation finale du mod√®le...\n")

# √âvaluation classique Keras
loss, acc = model.evaluate(test_generator)
print(f"‚úÖ Test Accuracy (Keras): {acc*100:.2f}%")
print(f"‚úÖ Test Loss: {loss:.4f}")

# Pr√©dictions sur le test set
y_pred = model.predict(test_generator)
y_pred_classes = np.argmax(y_pred, axis=1)
y_true = test_generator.classes

# Calcul des m√©triques
precision = precision_score(y_true, y_pred_classes, average='weighted')
recall = recall_score(y_true, y_pred_classes, average='weighted')
f1 = f1_score(y_true, y_pred_classes, average='weighted')
accuracy = accuracy_score(y_true, y_pred_classes)

# Affichage des r√©sultats
print("\nüìä M√©triques globales sur le jeu de test :")
print(f"   ‚Ä¢ Pr√©cision (Precision) : {precision*100:.2f}%")
print(f"   ‚Ä¢ Rappel (Recall)       : {recall*100:.2f}%")
print(f"   ‚Ä¢ F1-Score              : {f1*100:.2f}%")
print(f"   ‚Ä¢ Exactitude (Accuracy) : {accuracy*100:.2f}%")

# Rapport par classe
labels = list(test_generator.class_indices.keys())
report = classification_report(y_true, y_pred_classes, target_names=labels)
print("\nüìÑ Rapport d√©taill√© par classe :\n")
print(report)

# ============================================================
# 8Ô∏è‚É£ Courbe d‚Äôapprentissage ‚Äî Accuracy
# ============================================================
plt.figure(figsize=(8,6))
plt.plot(history.history['accuracy'], label='Entra√Ænement')
plt.plot(history.history['val_accuracy'], label='Validation')
plt.title("üìà Courbe de Pr√©cision (Accuracy)")
plt.xlabel("√âpoques")
plt.ylabel("Pr√©cision")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# 9Ô∏è‚É£ Courbe de la Loss
# ============================================================
plt.figure(figsize=(8,6))
plt.plot(history.history['loss'], label='Entra√Ænement')
plt.plot(history.history['val_loss'], label='Validation')
plt.title("üìâ Courbe de Perte (Loss)")
plt.xlabel("√âpoques")
plt.ylabel("Perte")
plt.legend()
plt.grid(True)
plt.show()

# ============================================================
# üîü Matrice de confusion
# ============================================================
cm = confusion_matrix(y_true, y_pred_classes)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=list(test_generator.class_indices.keys()))
plt.figure(figsize=(8,6))
disp.plot(cmap='Blues', xticks_rotation=45)
plt.title("üß© Matrice de Confusion")
plt.show()
```
